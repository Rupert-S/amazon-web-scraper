```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
URL = 'https://www.ebay.com/sch/i.html?_from=R40&_trksid=p4432023.m570.l1313&_nkw=car+shocks&_sacat=0'
```

```python
# Headers for request
HEADERS = ({
    'USER-AGENT': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36 Avast/127.0.0.0',
    'Accept-Language':'en-US, en;q=0.5'
           })
```

```python
#HTTP REQUEST
webpage = requests.get(URL, headers=HEADERS)
webpage
```

```python
type(webpage.content)
```

```python
soup = BeautifulSoup(webpage.content, "html.parser")
soup
```

```python
links = soup.find_all("a", attrs={'class':'s-item__link' })
```

```python
links
```

```python
product_list = links[2].get('href')
product_list
```

```python
new_webpage = requests.get(product_list, headers=HEADERS)
new_webpage
```

```python
new_soup = BeautifulSoup(new_webpage.content, "html.parser")
soup
```

```python
title = new_soup.find("h1", attrs = {'class' : 'x-item-title__mainTitle'}).text.strip()
title
```

```python
price = new_soup.find("div", attrs = {'class' : 'x-price-primary'}).text.strip()
price
```

```python
condition = new_soup.find("div", attrs = {'class' : 'x-item-condition-text'}).find("span", attrs={'class':'clipped'}).text.strip()
condition
```

```python
seller_rating = new_soup.find("div", attrs = {'class' : 'd-stores-info-categories__container__info__section__item'}).text.strip()
seller_rating
```

```python
seller_name = new_soup.find("h2", attrs = {'class' : 'd-stores-info-categories__container__info__section__title'}).text.strip()
seller_name
```

```python
placement = new_soup.find("dl", attrs = {'class' : 'ux-labels-values ux-labels-values--inline col-6 ux-labels-values--placementOnVehicle'}).find("div", attrs={'class':'ux-labels-values__values-content'}).text.strip()
placement
```

```python
items_included = new_soup.find("dl", attrs = {'class' : 'ux-labels-values ux-labels-values--inline col-6 ux-labels-values--itemsIncluded'}).find("div", attrs={'class':'ux-labels-values__values-content'}).text.strip()
items_included
```
